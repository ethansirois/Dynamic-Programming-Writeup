```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n_values = np.array([10, 18, 26, 34, 42, 50, 58, 66, 74, 82, 90, 98, 106])
rec_times = np.array([3.083e-06, 5.25e-06, 1.5792e-05, 0.000292417, 0.000740042, 0.00198788, 0.0232612, 0.042198, 0.0837106, 1.33687, 2.67482, 10.8211, 172.828])
dp_times = np.array([9.59e-07, 1.875e-06, 4.125e-06, 1.0208e-05, 1.7458e-05, 2.4792e-05, 5.125e-05, 5.8792e-05, 7.8709e-05, 0.000110875, 0.000139541, 0.000186417, 0.000254875])

plt.plot(n_values, rec_times, label = "Recursive Times") 
plt.plot(n_values, dp_times, label = "DP Times")
plt.xlabel("n")
plt.ylabel("time in seconds")
plt.title("Recursive vs DP Solution Runtimes")
plt.legend() 
plt.show()


plt.loglog(n_values, rec_times, label = "Recursive Times") 
plt.loglog(n_values, dp_times, label = "DP Times")
plt.xlabel("n")
plt.ylabel("time in seconds")
plt.title("LogLog Plot: Recursive vs DP Solution Runtimes")
plt.legend() 
plt.show()
```
